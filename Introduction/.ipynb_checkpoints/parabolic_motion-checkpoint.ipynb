# 出題内容
斜め上方へ投射された物体は放物線を描いて落下する。軌跡の長さが数十メートル程度であれば、地球の自転の影響などを考えずに済む。ここで、空気抵抗や地球の自転の影響を無視した上で、UTM座標系における物体の座標と速度ベクトルを求めたい。  

初速が250 m/sで仰角68.9°で打ち出された質量17.4 kgの物体の、落下までの座標と速度ベクトルを求めよ。時間ステップは0.05秒とする。求めた座標を基に、グラフを作成せよ。また、座標のグラフとは別に、時間ごとの速度のグラフも作成せよ。

鉛直上向きにy軸を取り、水平方向にx軸をとる

```python
# 解答例
import numpy as np
import matplotlib.pyplot as plt

v = 250    # 速度ベクトルの大きさ
v_arr = []
v_y = v * np.sin(68.9 / 180 * np.pi) # 速度ベクトル(y軸方向成分）
v_x = v * np.cos(68.9 / 180 * np.pi) # 速度ベクトル(x軸方向成分）
x, y = 0, 0                         # 初期座標
x_arr, y_arr = [], []               # 速度の時間変化をこの変数に記録
dt = 0.05                           # 微小時間[s]
count = 0
t = 0.0      # 時間[s]
t_arr = []   # 時間のリストを作成

while y >= 0 and count < 1000: # こういう計算では良くバグって無限ループになるので、必ず脱出する仕掛けを施す
    y += v_y * dt # 座標の計算
    x += v_x * dt
    y_arr.append(y)
    x_arr.append(x)
    v_arr.append((v_x**2 + v_y**2) ** 0.5) # 速度を計算を計算
    v_y += -9.8 * dt
    t_arr.append(t)
    t += dt
    count += 1

plt.plot(x_arr, y_arr) # プロットすると2次関数になっているので、解析解の特性と一致
plt.xlabel("x [m]")
plt.ylabel("y [m]")
plt.show()
```

```python
plt.plot(t_arr, v_arr)
plt.xlabel("time [s]")
plt.ylabel("velocity [m/s]")
plt.show()
```
