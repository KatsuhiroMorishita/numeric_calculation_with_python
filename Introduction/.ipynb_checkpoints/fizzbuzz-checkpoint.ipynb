# 出題内容
FizzBuzz問題は、プログラミングにおける基本的な問題です。これは複数人でやるゲームが基になっています。ゲームは1から順番にスタートし、3でも5でも割り切れる場合には"FizzBuzz"と答え、3で割り切れるが5では割り切れない場合には"Fizz"と答え、5では割り切れるが3では割り切れない場合には"Buzz"と答え、いずれでもない場合には数字を答えるというゲームです。答えを間違えると負けになります。

プログラムでこれを実現すると、条件分岐やfor文によるループ構造や関数など、プログラミングにおける基本を押さえることができます。FizzBuzz問題をPythonで書きなさい。

ちなみに、C言語で書くとこうなります。
<img src="http://www.korenlc.com/wp-content/uploads/2014/03/Screen-Shot-2014-03-20-at-1.29.25-AM1.png">

```python
# 解答例

def fizzbuzz(num):
    if num % 3 == 0 and num % 5 == 0: # 条件分岐
        return 'FizzBuzz'
    elif num % 3 == 0:
        return 'Fizz'
    elif num % 5 == 0:
        return 'Buzz'
    else:
        return str(num)

for num in range(1, 101):
    print(fizzbuzz(num))
    
```
