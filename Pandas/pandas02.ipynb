# pandas　AMeDASデータのグラフ描画
* amedas_47819.csvはAMeDAS熊本観測局のデータである
* 1時間当たりの降水量をグラフに表せ

データソース：  
[https://raw.githubusercontent.com/KatsuhiroMorishita/numeric_calculation_with_python/master/Pandas/amedas_47819.csv](https://raw.githubusercontent.com/KatsuhiroMorishita/numeric_calculation_with_python/master/Pandas/amedas_47819.csv)

```python
# 解答例
import pandas as pd

# csvファイルを読み込む。デリミタ（区切り文字が半角カンマなら引数のdelimiterは省略できる）
df = pd.read_csv("https://raw.githubusercontent.com/KatsuhiroMorishita/numeric_calculation_with_python/master/Pandas/amedas_47819.csv", delimiter=",", na_values=["--", "×"])
df   # 読み込んだDataFrameを表示（jupyterでのみ有効）
```

```python
# 型を確認（objectは文字列なので、そのままではグラフにできない）
df.dtypes
```

```python
# 日時を文字列から時刻オブジェクトへ変換
df["日時"] = pd.to_datetime(df["日時"])

# ついでに、雨量と気圧の数値化・・・（この例ではすでに数値化されているので実質的に変化はない）
df["降水量(mm)"] = df["降水量(mm)"].astype("float")
df["気圧(hPa)現地"] = df["気圧(hPa)現地"].astype("float")
```

```python
# 再度、型を確認（objectは文字列なので、そのままではグラフにできない）
df.dtypes
```

```python
# 日時と雨量の取り出し + ndarray型への変換
import numpy as np

times = np.array(df["日時"])
rains = np.array(df["降水量(mm)"])

# 確認のために表示
print(times)
print(rains)
```

```python
# グラフ描画
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(times, rains)
plt.show()
```

↑やはり、梅雨時に雨量が集中していることが分かる。
