```python
# 標準偏差の計算を試す
_std = np.std([1,2,3,4,5,6,6])
print("--1--", _std)

_std = np.std([1,2,3,4,5,6,6, float("nan")])   # 1つでもNaN（非値）が混じっていると、結果がNaNになってしまう
print("--2--", _std)
```

```python
#解答例
import numpy as np
import pandas as pd

def test(df):
    data = np.array(df["降水量(mm)"])
    #print(data)         # 降水量だけ抜き出せていることを確認したらコメントアウト
    y = np.std(data)
    return y

def main():
    df = pd.read_csv("https://raw.githubusercontent.com/KatsuhiroMorishita/numeric_calculation_with_python/master/Pandas/amedas_47819.csv", delimiter=",", na_values=["--", "×"])
    df = df.dropna(subset=["降水量(mm)"])   # 特定の列からNaNのあるレコードを削除する。左の例では、降水量が非値のレコードを削除した
    #print(df)         # 降水量の欄からNaNが消えていればOK。確認したらコメントアウト
    _std = test(df)
    print(_std)

if __name__ == "__main__":
    main()
```
