# 出題内容
引数で渡された値が奇数であれば文字列で”奇数”を返し、偶数であれば文字列で”偶数”を返す関数test()を作成せよ

```python
# ひな形
def test(x):
    return 1


def main():
    for i in range(100):
        print(i, test(i))
        
if __name__ == "__main__":
    main()
    
```

```python
# 解答例
def test(x):
    if x % 2 == 0:
        return "偶数"
    else:
        return "奇数"


def main():
    for i in range(100):
        print(i, test(i))
        
if __name__ == "__main__":
    main()
```
