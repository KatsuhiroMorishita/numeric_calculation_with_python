# fourier　AMeDASデータを使って、気圧の周波数特性を調べる
* amedas_47819.csvはAMeDAS熊本観測局のデータである
* 現地気圧の周波数成分を調べ、大きな順に3つの信号の周期[h]は何か求めよ

データソース：  
[https://raw.githubusercontent.com/KatsuhiroMorishita/numeric_calculation_with_python/master/Pandas/amedas_47819.csv](https://raw.githubusercontent.com/KatsuhiroMorishita/numeric_calculation_with_python/master/Pandas/amedas_47819.csv)

```python
# 解答例
import pandas as pd

# csvファイルを読み込む。デリミタ（区切り文字が半角カンマなら引数のdelimiterは省略できる）
#df = pd.read_csv("https://raw.githubusercontent.com/KatsuhiroMorishita/numeric_calculation_with_python/master/Pandas/amedas_47819.csv", delimiter=",", na_values=["--", "×"])
df = pd.read_csv("amedas_47819.csv", delimiter=",", na_values=["--", "×"])  # ローカルのファイルならこちら
print(df.dtypes)          # 各列のデータ型を確認
df                                  # 読み込んだDataFrameを表示（jupyterでのみ有効）
```

```python
# 日時を文字列から時刻オブジェクトへ変換
df["日時"] = pd.to_datetime(df["日時"])
```

```python
# 気圧の変化をグラフに表すことで値の変化を確認する
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = np.array(df["日時"])
y = np.array(df["気圧(hPa)現地"])
plt.plot(x, y)
plt.show()
```

↑何本か下に伸びている髭は台風の通過などを表している。北海道などでは寒冷前線が通過して（冷たいので密度の高い大気が侵入してきて）短時間に数hPa上昇することもある。

```python
# FFTを試みるが、上手くいかない
fft_size = 2**13    # len(y) > 2^nの中で、最大のnは13
s = y[:fft_size]
amp = np.fft.fft(s)    # フーリエ変換を実施
amp                             # フーリエ変換結果を表示（セルの最後に変数を書くと表示される機能は、jupyterでのみ有効）
```

↑ここではsの要素の中にnanが入っているのが駄目なポイントである。
nanをうまく処理する必要がある。

```python
# nanをnanになる前の数値で置き換える

_v = 0                            # 初期値
for i in range(len(y)):
    v = y[i]
    if np.isnan(v):           # 値がnanかどうか確認するには、isnan()という関数を使うしかない
        print("要素{0}番の値{1}を{2}に置換します".format(i, v, _v))
        y[i] = _v
        continue
    _v = v                        # nanでなければ、数値を代入して次のループへ
y
```

```python
# もう一度FFTを試みると、上手くいく
fft_size = 2**13    # len(y) > 2^nの中で、最大のnは13
s = y[:fft_size]
amp = np.fft.fft(s)    # フーリエ変換を実施
amp                             # フーリエ変換結果を表示（セルの最後に変数を書くと表示される機能は、jupyterでのみ有効）
```

```python
# グラフ描画を試みる（グラフは拡大ができるように、matplotlib notebookを指定する。これは、jupyterでのみ有効）
%matplotlib notebook
Ts = 3600   # サンプリング周期[s]
fs = 1 / Ts   # サンプリング周波数[Hz]

amp_half = amp[:int(len(amp) / 2)]       # フーリエ変換結果を半分に切断
freq = np.arange(0, fs/2, fs/fft_size)    # 横軸の数列を作成
plt.plot(freq, np.abs(amp_half))
plt.yscale("log")                                        # 縦軸の縮尺に差がありすぎるので、縦対数グラフとする
plt.grid()
plt.xlabel("frequency [Hz]")
plt.ylabel("amp.")
plt.show()
```

↑一部を拡大したグラフ

```python
# 周波数から周期を計算
f1 = 2.306e-5            # 成分の大きな周波数3つ
f2 = 1.149e-5
f3 = 3.463e-5

T1 = 1 / f1 / 3600    # 第１ピークの周期[h]
T2 = 1 / f2 / 3600    # 第2ピークの周期[h]
T3 = 1 / f3 / 3600    # 第3ピークの周期[h]

print(T1)
print(T2)
print(T3)
```

↑周波数から周期を計算してみると、太陽の周期とそれによって出来る定在波の周期であることが分かる。
大気は太陽によって加熱されて膨張し、圧力が大気を伝搬する。これを大気潮汐という。
