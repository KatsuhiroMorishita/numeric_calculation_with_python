```python
# (1)„ÅÆËß£Á≠î‰æã
import numpy as np
import matplotlib.pyplot as plt

# jupyter notebook„Åß„Éé„Éº„ÉàÂÜÖ„Å´„Ç∞„É©„Éï„ÇíÊèèÁîª„Åï„Åõ„ÇãÁâπÊÆäÂëΩ‰ª§
%matplotlib inline

def func(x):
    """ Âºè(1)„ÅÆË®àÁÆóÁµêÊûú„ÇíËøî„Åô
    """
    return (3.9**x) * (np.sin(x)**3)

def integral(f, x1, x2, dx): # ÂÆöÁ©çÂàÜÔºàÊ±éÁî®Ôºâ
    """ x1„Åã„Çâx2„Åæ„Åß„ÅÆÂÆöÁ©çÂàÜ„ÇíËøî„Åô
    f: Á©çÂàÜ„Åó„Åü„ÅÑÈñ¢Êï∞
    dx: x„ÅÆÂæÆÂ∞èÂ§âÂåñÈáè
    """
    x = np.arange(x1, x2, dx)
    ds = (f(x + dx) + f(x)) / 2.0 * dx # Âè∞ÂΩ¢Á©çÂàÜ„Åô„Çã„Å§„ÇÇ„Çä„Åß„ÄÅ„Åù„ÅÆÂæÆÂ∞èÈù¢Á©çds„ÇíÊ±Ç„ÇÅ„Çã
    s = np.sum(ds) # Èù¢Á©ç
    return s


def main():  
    x1, x2, dx = -10, 10, 0.01 # Á©çÂàÜÁØÑÂõ≤„Å®„ÄÅÂæÆÂ∞èÈáè
    print("Á©çÂàÜÂÄ§„ÅØ„ÄÅ", integral(func, x1, x2, dx))
    
    # „Ç∞„É©„Éï‰ΩúÊàê
    x = np.arange(x1, x2, dx)
    y = func(x)
    plt.plot(x, y)
    plt.grid()
    plt.show()

        
if __name__ == "__main__":
    main()

```

```python
# (2)„ÅÆËß£Á≠î‰æã
import numpy as np
import matplotlib.pyplot as plt

# jupyter notebook„Åß„Éé„Éº„ÉàÂÜÖ„Å´„Ç∞„É©„Éï„ÇíÊèèÁîª„Åï„Åõ„ÇãÁâπÊÆäÂëΩ‰ª§
%matplotlib inline

a, b, c = 5, -6, -20 # ‰øÇÊï∞„ÅØ„ÉÜ„Ç≠„Éà„Éº„Å´Ê±∫„ÇÅ„Åü

def func(x):
    """ Âºè(2)„ÅÆË®àÁÆóÁµêÊûú„ÇíËøî„Åô
    """
    return a * x**2 + b * x - c  # Èñ¢Êï∞ÂÜÖ„Å´ÂÆ£Ë®Ä„Åï„Çå„Å¶„ÅÑ„Å™„ÅÑÂ§âÊï∞„ÅØ„Åù„ÅÆÂ§ñ„ÅÆ„ÇÇ„ÅÆ„Åå‰Ωø„Çè„Çå„Çã

def integral(f, x1, x2, dx): # ÂÆöÁ©çÂàÜÔºàÊ±éÁî®Ôºâ
    """ x1„Åã„Çâx2„Åæ„Åß„ÅÆÂÆöÁ©çÂàÜ„ÇíËøî„Åô
    f: Á©çÂàÜ„Åó„Åü„ÅÑÈñ¢Êï∞
    dx: x„ÅÆÂæÆÂ∞èÂ§âÂåñÈáè
    """
    x = np.arange(x1, x2, dx)
    ds = (f(x + dx) + f(x)) / 2.0 * dx # Âè∞ÂΩ¢Á©çÂàÜ„Åô„Çã„Å§„ÇÇ„Çä„Åß„ÄÅ„Åù„ÅÆÂæÆÂ∞èÈù¢Á©çds„ÇíÊ±Ç„ÇÅ„Çã
    s = np.sum(ds) # Èù¢Á©ç
    return s


def main():  
    x1, x2, dx = -10, 10, 0.01 # Á©çÂàÜÁØÑÂõ≤„Å®„ÄÅÂæÆÂ∞èÈáè
    print("Á©çÂàÜÂÄ§„ÅØ„ÄÅ", integral(func, x1, x2, dx))
    
    # „Ç∞„É©„Éï‰ΩúÊàê
    x = np.arange(x1, x2, dx)
    y = func(x)
    plt.plot(x, y)
    plt.grid()
    plt.show()

        
if __name__ == "__main__":
    main()
```
