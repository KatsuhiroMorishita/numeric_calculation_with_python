```python
# „Åæ„Åö„ÄÅÈñ¢Êï∞„ÅÆ„Ç∞„É©„Éï„ÇíÁ¢∫Ë™ç
import numpy as np
import matplotlib.pyplot as plt

# ‚Üìjupyter‰∏ä„Åß„Ç∞„É©„Éï„Çí„Ç∫„Éº„É†„Åô„Çã„Åü„ÇÅ„ÅÆÂëΩ‰ª§
%matplotlib notebook

def f(x):
    return 2 * np.sin(3*x) - 1/7

x = np.arange(-5, 5, 0.01)
y = f(x)

plt.plot(x, y)
plt.grid()
plt.show()
```

```python
# Ëß£Á≠î‰æã
import numpy as np
import matplotlib.pyplot as plt

def diff(func, x, dx): # func„ÅØÂæÆÂàÜ„Åó„Åü„ÅÑÊñπÁ®ãÂºè„ÅÆÂÄ§„ÇíËøî„ÅôÈñ¢Êï∞
    return (func(x + dx) - func(x)) / dx

def calc_newton(func, x_init, dx): # Ê±éÁî®„Éã„É•„Éº„Éà„É≥Ê≥ï„ÅßËß£„ÇíÊ±Ç„ÇÅ„ÇãÈñ¢Êï∞
    x = x_init
    y = float("inf")
    count = 0
    e = 1.0

    while abs(y) > 0.01 and count < 1000 and e > 0.00001:
        y = func(x)
        dydx = diff(func, x, dx)
        new_x = -y / dydx + x
        e = abs(new_x - x) / (abs(x) + 0.0000000001)
        x = new_x
        count += 1

    print(x, count) # ‰Ωø„ÅÜÊôÇ„Å´„ÅØ„Ç≥„É°„É≥„Éà„Ç¢„Ç¶„Éà„Åó„ÅüÊñπ„ÅåËâØ„ÅÑ
    return x


def main():
    pos_x = calc_newton(f, 4, 0.01) # ‰∫§ÁÇπ„ÅÆxÂ∫ßÊ®ô„ÇíÊ±Ç„ÇÅ„Çã„ÄÇÈñ¢Êï∞f()„ÅØ‰∏ä„ÅÆ„Çª„É´„ÅßÂÆöÁæ©„Åó„Åü„ÇÇ„ÅÆ„Çí‰Ωø„ÅÜÔºà‰∏ä„ÅÆ„Çª„É´„ÇíÂÖà„Å´ÂÆüË°å„Åô„Çã„Åì„Å®Ôºâ
    print("y = 0„Å®„ÅÆ‰∫§ÁÇπx„ÅØ„ÄÅ", pos_x)
        
if __name__ == "__main__":
    main()
```
